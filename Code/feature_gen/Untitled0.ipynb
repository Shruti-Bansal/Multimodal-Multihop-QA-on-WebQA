```python
import base64
from PIL import Image
from io import BytesIO
from tqdm import tqdm

failed_ids = []
# extract
x=0
with open("/content/drive/MyDrive/imgs.tsv", "r") as f:
    for line in tqdm(f):
            x+=1
            if x==64000:
                break
            try:
                img_id, img = line.strip().split('\t')
                img = Image.open(BytesIO(base64.b64decode(img)))
                img = img.convert('RGB')
                img.save('/content/drive/MyDrive/WebQA_x101fpn/images/'+img_id+'.jpg')
            except:
                    failed_ids.append(img_id)
print(len(failed_ids))
```
