```python
import os

# Specify the range of folder names
start_folder = 65
end_folder = 139

# Specify the directory where you want to create the folders
base_directory = "/home/ubuntu/Project/11-777-Project-aa/Data/test"

# Create folders
for folder_name in range(start_folder, end_folder + 1):
    folder_path = os.path.join(base_directory, str(folder_name))
    
    # Check if the folder doesn't exist, then create it
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder '{folder_name}' created at '{folder_path}'")
    else:
        print(f"Folder '{folder_name}' already exists at '{folder_path}'")
```

```python
def decide_destination_path(source_path):
    """
    Decide the destination path for a file based on a custom rule.
    
    Args:
        source_path (str): The path of the source file.

    Returns:
        str: The path of the destination directory.
    """

    imgid=int(source_path.split('/')[-1].split('.')[0])
    f_no=(imgid%10000000)//1000
    if f_no>65 and f_no<139:
        destination_path=os.path.join("/home/ubuntu/Project/11-777-Project-aa/Data/test/",str(f_no))
    else:
        destination_path = os.path.join("/home/ubuntu/Project/11-777-Project-aa/Data/train/",str(f_no))



    return destination_path

# Example usage:
source_file_path = "/home/ubuntu/Project/11-777-Project-aa/Data/content/drive/MyDrive/WebQA_x101fpn/output/sushil_data2/40065039.pkl"
destination_path = decide_destination_path(source_file_path)
print(f"Destination path for '{source_file_path}': {destination_path}")
```

```python
import os
import shutil

# Specify source and destination directories
source_directory = "/home/ubuntu/Project/11-777-Project-aa/Data/content/drive/MyDrive/WebQA_x101fpn/output/sushil_data2"


# Define the rule (example: move files with '.txt' extension)
file_extension_rule = '.pkl'

# Iterate through the files in the source directory
for filename in os.listdir(source_directory):
    source_path = os.path.join(source_directory, filename)

    # Check if the item is a file and meets the rule
    if os.path.isfile(source_path) and filename.endswith(file_extension_rule):
        destination_path = decide_destination_path(source_path)

        # Move the file to the destination directory
        shutil.move(source_path, destination_path)
        print(f"File '{filename}' moved to '{destination_path}'")


```
